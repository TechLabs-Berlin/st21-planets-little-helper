```python
plh_df.rename(columns= {'In your opinion in which categories you can improve your lifestyle to be more environmentally friendly: (Choose one or more)': 'Category',     'How much time are you willing to spend on taking more environmentally friendly actions?':'Duration',  'How much money are you willing to spend on achieving your sustainable goals?': 'Budget',  'What gender do you identify as?':'Sex', 'What is your age?':'Age' }, inplace=True )
```

```python
plh_df.drop(['I consider the potential environmental impact of my actions when making many of my decisions.',
       'I am concerned about wasting the resources of our planet.',
       'I feel like it’s sometimes complicated to live a sustainable lifestyle.',
       'I am willing to be inconvenienced in order to take actions that are more environmentally friendly.',
       'Where do you get knowledge about environmental issues from? (Choose one or more)', 'Do you use any technological solutions (e.g. apps) to support your green lifestyle? ',
       'Can you list some of those solutions? (Optional)',
       'Will you consider using an app that can assist you with your sustainable goals?',
       'What will be your main reason for using this kind of app? (Optional)', 'What is the highest degree or level of education you have completed?',
       'Email address for further questions or product testing. (Optional)'], axis=1, inplace=True)
```

```python
plh_df.drop([0,2], axis=0, inplace=True)
plh_df.isnull().sum()
```

```python
plh_df.loc[(plh_df['Duration'].isnull()==True), 'Duration']='Every day'
plh_df
```

```python
plh_df.loc[(plh_df['Sex'].isnull()==True), 'Sex']='Prefer not to answer'
plh_df.loc[(plh_df['Category'].isnull()==True), 'Category']='Food, Energy, Recycling, Transport and travel'
plh_df.loc[(plh_df['Budget'].isnull()==True), 'Budget']='None'
plh_df
```

```python
plh_df['Category'].replace(['Transport and travel, Energy, Home and work, recycling'],['Transport and travel, Energy, Home and work, Recycling'],inplace=True)
plh_df.loc[:, 'Category']
```

```python
print(pd.get_dummies(plh_df['Category']).head(5))
plh_df['Category_list'] = plh_df.Category.str.split(', ')
pln = pd.get_dummies(plh_df['Category_list'].apply(pd.Series).stack()).sum(level=0)
pln.columns
```

```python
pln.drop(["Not sure if you mean it in a general way or the question is in what areas could I do even more improvements on top of what I've been doing previously?", 'Shopping','also dairy products. If it would be in ordinary supermarkets like Lidl or Edeka.', 'bureaucracy', 'cereals and legumes', 'clothes',
       'food: especially no one way plastic package for pasta',  'use less plastic'], axis=1, inplace=True)
pln.drop(['Advocacy','Home and work'], axis=1, inplace=True)
plhpln_df = pd.merge(plh_df, pln, left_index=True, right_index=True)
plhpln_df
```

```python
plhpln_df.drop(['Category'], axis=1, inplace=True)
dmy = pd.get_dummies(plhpln_df, columns=['Duration','Budget','Sex','Age'])
dmy
```

```python
titles =['Timestamp',
 'Duration_A few times a month',
 'Duration_A few times a week',
 'Duration_About once a week',
 'Duration_Every day',
 'Duration_Less than once a month',
 'Duration_Once a month',
 'Budget_1-10€',
 'Budget_10-100€',
 'Budget_>100€',
 'Budget_No difference',
 'Budget_None',
 'Sex_Female',
 'Sex_Male',
 'Sex_Prefer not to answer',
 'Age_18-24',
 'Age_25-34',
 'Age_35-44',
 'Age_45-54',
 'Age_55-64',
 'Age_Over 65',
 'Energy',
 'Food',
 'Recycling',
 'Transport and travel','Category_list']
titles
```

```python
dmy = dmy[titles]
dmy.drop(['Category_list', 'Timestamp'], axis=1, inplace=True)
dmy
```

```python
dmy.describe()
```

```python
correlations = dmy.corr()
f,ax=plt.subplots(figsize=(15,10))
sns.heatmap(correlations,annot=True)
```

```python
#kmeans algorithm
#1 Elbow

#scale data
scaler=StandardScaler()
dmy_scaled = scaler.fit_transform(dmy)
```

```python
dmy_scaled.shape
```

```python
dmy_scaled
```

```python
scores=[]

range_values=range(1,20)

for i in range_values:
  kmean = KMeans(n_clusters=i)
  kmean.fit(dmy_scaled)
  scores.append(kmean.inertia_)

plt.plot(scores,'bx-')
```

```python
#apply kmeans

kmeans = KMeans(4)
kmeans.fit(dmy_scaled)

label = kmeans.labels_
```

```python
kmeans.cluster_centers_.shape
```

```python
cluster_centers = pd.DataFrame(data=kmeans.cluster_centers_,columns = [dmy.columns])
cluster_centers
```

```python
#inversed transform
cluster_centers = scaler.inverse_transform(cluster_centers)

cluster_centers = pd.DataFrame(data=cluster_centers, columns=[dmy.columns])

cluster_centers
```

```python
label.shape
```

```python
#predict

y = kmeans.fit_predict(dmy_scaled)
y
```

```python
dmy_cluster = pd.concat([dmy, pd.DataFrame({'cluster':label})],axis=1)

dmy_cluster
```

```python
#principal component analysis
  #dimensionality reduction

pca = PCA(n_components=2)
principal_comp = pca.fit_transform(dmy_scaled)
principal_comp
```

```python
pca_df = pd.DataFrame(data=principal_comp,columns=['pca1','pca2'])
pca_df
```

```python
pca_df = pd.concat([pca_df,pd.DataFrame({'cluster':label})],axis=1)
pca_df
```

```python
plt.figure(figsize=(5,5))
ax= sns.scatterplot(x='pca1', y='pca2', hue='cluster',data=pca_df,palette=['red','green','blue','yellow'])
plt.show()
```
